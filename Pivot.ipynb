```python
import pandas as pd
```

```python
df1=pd.DataFrame({"Names":["Abdul","Rahiman","Vamsi","Satya","Srikanth"]
                  ,"Age":[24,23,34,54,35],
                  "salary":[24500,23000,45000,53333,45555],
                  "Company":["TCS","TCS","Infosys","Wipro","Wipro"],
                 "city":["Banglore","Banglore","Banglore","Banglore","Banglore"]})
```

```python
df1
```

```python
df1.pivot(index="Names",columns="city")
```

```python
#values---if we only want particular columns then we can give required colums in values
df1.pivot(index="Names",columns="city",values=["Age","Company"])
```

```python
#pivot_table--gives the summary
```

```python
df1.pivot_table(index="Company",columns="city",aggfunc="max",values=["Age","salary"])
```
